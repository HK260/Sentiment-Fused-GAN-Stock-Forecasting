```python
import os
import numpy as np
import csv
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import math
```

```python
df = pd.read_csv("Reliance_Stock_2020_2024.csv",parse_dates=['date'])
```

```python
print(df.head())
print(df.shape)
print(df.columns)
```

```python
df.rename(columns={'1. open': 'Open', '2. high': 'High', '3. low': 'Low', '4. close': 'Close', '5. volume': 'Volume'}, inplace=True)
```

```python
print(df.head())
print(df.shape)
print(df.columns)
```

```python
fig, ax = plt.subplots(figsize=(10, 3))
ax.plot(df['date'], df['Close'], label='Reliance stock')
ax.set(xlabel="Date", ylabel="Rupee", title="Reliance Stock Price")
date_form = DateFormatter("%Y")
ax.xaxis.set_major_formatter(date_form)
plt.show()
```

**Reliance Industries Limited (RIL) announced a 1:1 bonus share issue in 2024, which likely explains the sharp drop observed in the unadjusted stock price. Therefore, data from 2024 onwards will be excluded from analysis, as it introduces structural discontinuity and acts as an outlier. This exclusion also helps in optimizing computational efficiency.**


```python
df = df[(df["date"] >= "2020-01-01") & (df["date"] <= "2023-12-31")]
```

```python
fig, ax = plt.subplots(figsize=(10, 3))
ax.plot(df['date'], df['Close'], label='Reliance stock')
ax.set(xlabel="Date", ylabel="Rupee", title="Reliance Stock Price")
date_form = DateFormatter("%Y")
ax.xaxis.set_major_formatter(date_form)
plt.show()
```

```python
# Set 'Date' as the index and sort in ascending order
df = df.set_index('date').sort_index(ascending=True)
```

```python
df.head()
```

```python
# Calculate technical indicators
def get_technical_indicators(data):
    # Create 7 and 21 days Moving Average
    data['MA7'] = data.iloc[:,3].rolling(window=7).mean()
    data['MA21'] = data.iloc[:,3].rolling(window=21).mean()

    # Create Exponential moving average
    data['EMA'] = data.iloc[:,3].ewm(com=0.5).mean()

    return data
T_df = get_technical_indicators(df)
```

```python
T_df.to_csv("Final_preprocessed_stock_data.csv", index=True)
```

```python
def plot_technical_indicators_full(dataset):
    plt.figure(figsize=(16, 6), dpi=100)
    plt.plot(dataset.index, dataset['MA7'], label='MA 7', color='green', linestyle='--', linewidth=1.2)
    plt.plot(dataset.index, dataset['MA21'], label='MA 21', color='red', linestyle='--', linewidth=1.2)
    plt.plot(dataset.index, dataset['EMA'], label='EMA', color='purple', linestyle='-.', linewidth=1.2)

    plt.title('Technical Indicators - Full Dataset')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()
```

```python
plot_technical_indicators_full(T_df)
```
