## 1.a) Stock Data Fetching & Caching (API Calls)

### In this step we are fetching and caching stock data of Reliance Industries. 

---

<details>
  <summary><strong>üì° 1. Purpose & Overview</strong></summary>

<br>

This cell fetches historical stock data for **Reliance Industries** from the **Alpha Vantage API**, but does so intelligently.

**‚úÖ Key Features:**

- Checks if the CSV already exists locally  
- Skips API call if the file is present, non-empty, and contains up-to-date data  
- Automatically filters data between `2020-01-01` and `2024-12-31`  
- Saves the clean data to a CSV file  
- Prints summary statistics and a quick preview of the dataset  

</details>

---

<details>
  <summary><strong>üß† 2. Why This Matters</strong></summary>

<br>

Fetching data from the web **every time** can lead to:

- ‚ùå Wasted bandwidth  
- ‚ùå Slower execution  
- ‚ùå Unnecessary consumption of limited API credits  

**‚úÖ This logic helps** in reusing saved data when it's valid, and only re-fetch when necessary (e.g., if the file is missing, empty, or stale).

</details>

---

<details>
  <summary><strong>üìé 3. Output & Results</strong></summary>

<br>

**Saved CSV**:  
`Reliance_Stock_2020_2024.csv`

**Summary Statistics**:  
- Displayed using `.describe()` on the filtered data

**Preview**:  
- First 5 rows: `.head()`  
- Last 5 rows: `.tail()`  

</details>

---

```python
import os
import pandas as pd
from datetime import datetime, timedelta
from alpha_vantage.timeseries import TimeSeries

# === Configuration ===
API_KEY = "L2XYTAMN2JEBGIZM"
symbol = "RELIANCE.BSE"
output_file = "Reliance_Stock_2020_2024.csv"
start_date = "2020-01-01"
end_date = "2024-12-31"

# === Helper function to check file freshness ===
def is_data_stale(file_path, expected_end_date):
    try:
        df = pd.read_csv(file_path, index_col=0, parse_dates=True)
        if df.empty:
            return True
        last_date = df.index.max()
        return last_date < pd.to_datetime(expected_end_date)
    except Exception as e:
        print(f"‚ö†Ô∏è Error reading existing file: {e}")
        return True

# === Fetch from API only if needed ===
needs_fetch = not os.path.exists(output_file) or is_data_stale(output_file, end_date)

if needs_fetch:
    print(f"üì° Fetching fresh stock data for {symbol} from Alpha Vantage...")
    ts = TimeSeries(key=API_KEY, output_format="pandas")

    try:
        data, meta_data = ts.get_daily(symbol=symbol, outputsize="full")
        data.index = pd.to_datetime(data.index)
        data_filtered = data[(data.index >= start_date) & (data.index <= end_date)]
        data_filtered.to_csv(output_file)
        print(f"‚úÖ Stock data saved to: {output_file}")
    except Exception as e:
        print(f"‚ùå Error fetching or saving data: {e}")
        exit(1)
else:
    print(f"üìÅ Using cached data from: {output_file}")
    data_filtered = pd.read_csv(output_file, index_col=0, parse_dates=True)

# === Summary and samples ===
print("\nüìä Data Summary:")
print(data_filtered.describe())

print("\nüîπ First 5 rows:")
print(data_filtered.head())

print("\nüîπ Last 5 rows:")
print(data_filtered.tail())
```

## 1.b) News Headline Collection (RSS Feed)
### In this step we are collecting monthly news headlines of Reliance Industries 
---

<details> 
  <summary><strong>üì∞ 1. Purpose & Overview</strong></summary> 
  <br>

This section scrapes monthly news headlines related to Reliance Industries between Jan 2020 and Dec 2024 using Google News RSS feeds.

**‚úÖ Key Features:**

- Uses the `feedparser` library to read RSS feeds  
- Constructs URLs that fetch India-specific news in English  
- Iterates month-by-month (due to a 100-article limit per request)  
- Collects article titles and published dates  
- Saves results into a single CSV file, **but only downloads fresh data if the file is missing or stale**  
- Adds basic caching to avoid unnecessary downloads

</details>

---

<details> 
  <summary><strong>üß† 2. Why This Matters</strong></summary> 
  <br>

News headlines often contain sentiment, event-driven patterns, and context that can influence stock behavior.

**‚úÖ Advantages of this approach:**

- Avoids duplication and quota issues by limiting each query to one month  
- Reduces redundant data fetching by checking for stale or missing files  
- Enables time-aligned analysis with stock data  
- Easy to preprocess and integrate with natural language models  
- Allows for monthly aggregation, frequency analysis, and sentiment tracking  

</details>

---

<details> 
  <summary><strong>üìé 3. Output & Results</strong></summary> 
  <br>

**Saved CSV**:  
`Reliance_GoogleNews_Monthly_2020_2024.csv`

**Columns**:

- `Headline` ‚Äî The news headline text  
- `Published Date` ‚Äî Date the article was published  

**Sample Preview**:
- Displayed using `.head()` on the final DataFrame  
- **Data is not pre-filtered or cleaned yet** ‚Äî this will be handled in the preprocessing stage  

**‚úÖ Total Headlines Fetched**:
- Printed at the end using `len(df)`

**‚ÑπÔ∏è Note**:
- If a valid CSV already exists with data up to December 2024, it is reused instead of fetching again.

</details>

---


```python
import os
import pandas as pd
from datetime import datetime, timedelta
import feedparser

# === Configuration ===
output_file = "Reliance_GoogleNews_Monthly_2020_2024.csv"
start_year = 2020
end_year = 2024

# === Check if data file exists and is non-empty ===
def is_file_missing_or_empty(file_path):
    return not os.path.exists(file_path) or os.path.getsize(file_path) == 0

# === Function to fetch RSS news for a specific month ===
def fetch_monthly_news(year, month):
    start_date = f"{year}-{month:02d}-01"
    end_date = (datetime(year, month, 1) + timedelta(days=31)).replace(day=1) - timedelta(days=1)
    end_date_str = end_date.strftime("%Y-%m-%d")

    rss_url = (
        f"https://news.google.com/rss/search?q=Reliance+Industries"
        f"+after:{start_date}+before:{end_date_str}&hl=en-IN&gl=IN&ceid=IN:en"
    )

    feed = feedparser.parse(rss_url)
    headlines = []

    for entry in feed.entries:
        headlines.append({
            "Headline": entry.title,
            "Published Date": getattr(entry, "published", start_date)
        })

    return headlines

# === Main logic with caching ===
if is_file_missing_or_empty(output_file):
    print("üì° Fetching monthly news headlines from Google News RSS...")
    all_news = []

    for year in range(start_year, end_year + 1):
        for month in range(1, 13):
            print(f"üìÖ Fetching: {year}-{month:02d}")
            monthly_headlines = fetch_monthly_news(year, month)
            all_news.extend(monthly_headlines)

    # Save to CSV
    df = pd.DataFrame(all_news)
    df.to_csv(output_file, index=False)
    print(f"\n‚úÖ Saved {len(df)} headlines to {output_file}")
else:
    print(f"üìÅ Using cached headlines from: {output_file}")
    df = pd.read_csv(output_file)

# === Summary ===
print("\nüìä News Headlines Summary:")
print(f"üìù Total headlines collected: {len(df)}")

try:
    print(f"üìÖ Date range: {df['Published Date'].min()} to {df['Published Date'].max()}")
except Exception as e:
    print(f"‚ö†Ô∏è Unable to determine date range: {e}")

print("\nüîπ First 5 headlines:")
print(df.head()[['Published Date', 'Headline']])

print("\nüîπ Last 5 headlines:")
print(df.tail()[['Published Date', 'Headline']])
```

# 2. Data Preprocessing Stage
## 2.a) News Data Preprocessing & Monthly Aggregation
### In this step, we clean and preprocess Google News articles for Reliance Industries 

---

<details>
  <summary><strong>üì° 1. Purpose & Overview</strong></summary>

<br>

This cell performs preprocessing on a dataset of Google News articles about Reliance Industries between 2020 and 2024.

**‚úÖ Key Features:**

- Loads the raw dataset from a CSV  
- Converts the `'Published Date'` to a clean `YYYY/MM/DD` format  
- Drops missing dates and duplicates  
- Randomly selects one article per day  
- Aggregates monthly article counts  
- Saves the cleaned dataset to a new CSV  
- Visualizes monthly distribution with a line chart  

</details>

---

<details>
  <summary><strong>üß† 2. Why This Matters</strong></summary>

<br>

News datasets often contain:

- üóìÔ∏è Duplicate entries per day  
- ‚ùå Missing or malformed date formats  
- üîÅ Redundant information

**‚úÖ This logic helps** ensure we get a balanced, one-news-per-day dataset for further analysis (e.g., sentiment tracking). Grouping by month allows us to visualize and validate data consistency over time.

</details>

---

<details> 
  <summary><strong>üìé 3. Output & Results</strong></summary> 
  <br>

**Saved CSV**:  
`Reliance_GoogleNews_Monthly_2020_2024_preprocessed.csv`

**Columns**:

- `Headline` ‚Äî The news headline text  
- `Published Date` ‚Äî Date in `YYYY/MM/DD` format  
- `Year_Month` ‚Äî Aggregated format used for monthly grouping (e.g., `2021/03`)  

**Sample Preview**:
- Displayed using `.head()` on the final DataFrame  
- Shows one randomly selected article per day (after cleaning & deduplication)

**‚úÖ Total Headlines Used**:
- Printed at the end using `len(df)`

**üìä Monthly News Count**:
- Aggregated using `groupby("Year_Month")`  
- Visualized using a line chart (`matplotlib`)

</details>

---

```python
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt

# Load the saved dataset
df = pd.read_csv("Reliance_GoogleNews_Monthly_2020_2024.csv")

# Convert 'Published Date' to datetime and format to YYYY/MM/DD
df["Published Date"] = pd.to_datetime(
    df["Published Date"], format="%a, %d %b %Y %H:%M:%S GMT"
).dt.strftime("%Y/%m/%d")

# Drop rows with missing Published Date if any
df.dropna(subset=["Published Date"], inplace=True)

# Shuffle and drop duplicates (one news per day)
df = df.sample(frac=1, random_state=42)
df.drop_duplicates(subset=["Published Date"], inplace=True)
df.reset_index(drop=True, inplace=True)

print(f"‚úÖ Randomly selected one article per day. Remaining: {len(df)}")

# Extract Year-Month (YYYY/MM)
df["Year_Month"] = df["Published Date"].str.slice(0, 7)

# Count number of articles per month
monthly_counts = df.groupby("Year_Month").size().reset_index(name="News Count")

# Sort by date
monthly_counts = monthly_counts.sort_values("Year_Month").reset_index(drop=True)

# Display first few rows of monthly article counts
print("üìä News articles per month:\n", monthly_counts.head())

# Save the cleaned dataset
df.to_csv("Reliance_GoogleNews_Monthly_2020_2024_preprocessed.csv", index=False)

# Optional: Plot monthly news article count
plt.figure(figsize=(12, 6))
plt.plot(monthly_counts["Year_Month"], monthly_counts["News Count"], marker='o')
plt.xticks(rotation=45)
plt.xlabel("Year-Month")
plt.ylabel("News Count")
plt.title("Monthly News Article Count (Reliance, 2020-2024)")
plt.grid(True)
plt.tight_layout()
plt.show()

```
